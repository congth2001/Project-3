```python
import pandas as pd
```

```python
df_players=pd.read_csv('archive/players.csv')
df_valuations=pd.read_csv('archive/player_valuations.csv')
df_clubs=pd.read_csv('archive/clubs.csv')
df_competitions=pd.read_csv('archive/competitions.csv')
df_clubs
```

```python
merged_1=pd.merge(df_valuations, df_players, how='left',on=['player_id'])
merged_1=merged_1.iloc[:,:8]
merged_1.columns=['date', 'datetime','dateweek', 'player_id','club_id','market_value_in_eur','domestic_competition_id','player_name']
merged_1

merged_2=pd.merge(merged_1, df_clubs, how='left', on=['club_id'])
merged_2=merged_2.iloc[:,[0,1,2,3,4,5,6,7,9]]
merged_2.columns=['date', 'datetime','dateweek', 'player_id','club_id','market_value_in_eur','competition_id','player_name','club_name']
merged_2

merged_3=pd.merge(merged_2, df_competitions, how='left', on=['competition_id'])
merged_3=merged_3.iloc[:,[0,3,4,5,6,7,8,10]]
merged_3.columns=['date', 'player_id','club_id','market_value_in_eur','competition_id','player_name','club_name', 'competition_name']
merged_3
```

```python
rs = merged_3[merged_3.competition_id.isin(['GB1','L1','ES1','IT1','FR1'])]
rs.to_csv('market_value.csv', index=None)
```

```python
special_char = ['\xad','À','Á','Ç','É','Ñ','Ó','Ö','Ø','Ü','ß',
                     'à','á','â','ã','ä','å','æ','ç','è','é','ê','ë',
                     'í','î','ï','ð','ñ','ò','ó','ô','ö','ø','ú','ü',
                     'ý','ă','ą','Ć','ć','Č','č','Đ','đ','ę','ě','ğ',
                     'İ','ı','ľ','Ł','ł','ń','ň','ō','ř','Ş','ş','Š',
                     'š','ů','Ž','ž','Ș','ș','ț']

replacement = ['','A','A','C','E','N','O','O','O','U','ss',
                'a','a','a','a','a','a','ae','c','e','e','e',
                'e','i','i','i','d','n','o','o','o','o','o',
                'u','u','y','a','a','C','c','C','c','D','d',
                'e','e','g','I','i','l','L','l','n','n','o',
                'r','S','s','S','s','u','Z','z','S','s','t']
```

```python
df_valuations=pd.read_csv('market_value_newest.csv')
```

```python
clean_players = []

for a in df_valuations['Player']:
  string = a

  player_string = []
  for z in string:
    if z in special_char:
      for e in list(range(len(special_char))):
        if z == special_char[e]:
          player_string.append(replacement[e])
        else:
          pass
    else:
      player_string.append(z)
  player_string = ''.join(player_string)
  clean_players.append(player_string)



df_valuations['Player'] = clean_players 
df_valuations.to_csv('market_value_newest.csv', index=None)
```
