```python
unique_player=df_overview['Player'].unique()
columns=['Tournament', 'Player', 'Number', 'Nation', 'Pos',
       'Age', 'Min', 'Performance_Gls', 'Performance_Ast', 'Performance_PK',
       'Performance_PKatt', 'Performance_Sh', 'Performance_SoT',
       'Performance_CrdY', 'Performance_CrdR', 'Performance_Touches',
       'Performance_Tkl', 'Performance_Int', 'Performance_Blocks',
       'Expected_xG', 'Expected_npxG', 'Expected_xAG', 'SCA', 'GCA',
       'Passes_Cmp', 'Passes_Att', 'Passes_Cmp%', 'Passes_Prog', 'Passes_Succ',
       'Dribbles_Att']
df_overview_mean = pd.DataFrame(columns=columns)

for player in unique_player:
  # player=unique_player[0]
  player_df =  df_overview[df_overview.Player.isin([player])]
  values=[]
  values.append(player_df.Tournament.values[0])
  values.append(player_df.Player.values[0])
  values.append(player_df.Number.values[0])
  values.append(player_df.Nation.values[0])
  values.append(player_df.Pos.values[0])
  values.append(player_df.Age.values[0].split('-')[0])
  mean = player_df.mean().values
  if len(mean) > 25:
    mean = np.delete(mean, 0)
  for i in mean:
    values.append(i)
  values.pop(6)
  

  df_overview_mean = pd.DataFrame(np.insert(df_overview_mean.values, 0, values=values, axis=0))
df_overview_mean.columns=columns
df_overview_mean

```

```python
df_overview_mean.to_csv('playerOverviewMean.csv', index=None)
```

```python
pos={'Defender':['CB','WB','LB','RB'],'Midfielder':['LM','CM','RM','DM','AM'],'Forward':['FW','LW','RW'],'Goal Keeper':['GK']}
key_list=list(pos.keys())
val_list=list(pos.values())
val_list
```

```python
position=[]
for cat in df.Pos.values:
  cat=cat.split(',')
  if len(cat)<=2:
    for val in val_list:
      for item in val:
        if cat[0]==item:
          cat=key_list[val_list.index(val)]
          # print(cat)
          break
  else:
    last={'B':0,'M':0,'W':0}
    for i in cat:
      if i[1]=='B': last['B']+=1
      elif i[1]=='M': last['M']+=1
      else: last['W']+=1
    # print(last)
    last=sorted(last.items(),key=lambda x:x[1], reverse=True)
    # print(last)
    # keys=list(last.keys())
    if last[0][0]=='B': cat='Defender'
    elif last[0][0]=='M': cat='Midfielder'
    elif last[0][0]=='W': cat='Forward'
    # print(cat)
  position.append(cat)
  
len(position)
```

```python
df.insert(8, 'Position', position, True)
df.to_csv('playerOverviewValuation.csv',index=None)
```
